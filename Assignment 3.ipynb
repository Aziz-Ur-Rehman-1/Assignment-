```python
#LAB TASK 5
sentence = input("Enter a sentence: ")
word = input("Enter word: ")

words = sentence.split()     
new_words = []
for w in words:
    if w != word:
        new_words.append(w)

result = ""
for i in range(len(new_words)):
    result = result + new_words[i]
    if i != len(new_words)-1:
        result = result + " "

print(result)

#Output
#Python is snake and i learn Python🐍
```

```python
# lab task 7

sentence = input("Enter a sentence: ")
words = sentence.split()
capitalized_words = []

for w in words:
    if len(w) > 0:
        first = w[0].upper()
        rest = w[1:].lower()
        capitalized_words.append(first + rest)
    else:
        capitalized_words.append(w)

result = ""
for i in range(len(capitalized_words)):
    result = result + capitalized_words[i]
    if i != len(capitalized_words)-1:
        result = result + " "

print(result)

#Output
#hello world
```
